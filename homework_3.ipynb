#### Задание 1

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd

from sklearn.datasets import load_boston
```

```python
boston = load_boston()
boston.keys()
```

```python
X = pd.DataFrame(boston["data"], columns=boston["feature_names"])
y = pd.DataFrame(boston["target"], columns=["price"])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

#### Задание 2

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
clf = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42)

clf.fit(X_train, y_train.values[:, 0])

y_pred = clf.predict(X_test)

r2_score(y_test, y_pred)
```

Метрика R2 у модели Random Forest выше, чем у регрессионной, соответственно данная модель работает лучше.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('creditcard.csv')
df
```

```python
df['Class'].value_counts(normalize=True)
```

```python
pd.options.display.max_columns = 100
df.info()
```

```python
df.head(10)
```

```python
X = pd.DataFrame(df.loc[:, df.columns != 'Class'])
X
```

```python
y = df['Class']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.3, random_state=100)
```

```python
X.shape
```

```python
y.shape
```

```python
parameters = {
'n_estimators': [10, 15],
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)
}

clf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=100),
    param_grid=parameters,
    scoring='roc_auc',
    cv=3,
)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
clf.predict_proba(X_test)
```

```python
y_pred_proba = clf.predict_proba(X_test)[:, 1]
y_pred_proba
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_pred_proba)
```
